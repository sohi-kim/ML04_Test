```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X=np.array([[1,1],[1,2],[2,2],[2,3]])
np.shape(X)
y=[6,8,9,11]
```

```python
#y=1*x0 +2*x1+3
y=np.dot(X, np.array([1,2])) + 3   # 행렬의 곱셈. 1과2는 두 개의 독립변수에 대한 가중치
```

```python
y
```

```python
# 모델링을 하고 x0=3, x1=5 일때 y의 예측값은?
model = LinearRegression()
reg = model.fit(X,y)     # 모델의 학습으로 모델이 생성
```

```python
reg.predict(np.array([[3,5]]))    # 예측하기
```

```python
reg.intercept_  # 절편 bias
```

```python
reg.coef_    # coefficient (계수) 기울기. 가중치 변수값
# 선형 회귀(linear regression) 모델에서 회귀 모델이 학습한 후,
# 독립 변수(independent variables)에 대한 각 가중치값을 구해준다.
```
